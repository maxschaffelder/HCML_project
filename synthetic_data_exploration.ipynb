```python
# Explore feature distributions for both datasets

# Function to create side-by-side bar plots for each feature
def plot_side_by_side(original_df, synthetic_df, feature):
    fig, axes = plt.subplots(1, 2, figsize=(18, 6), sharey=True)
    
    # Calculate percentages for original data
    original_counts = original_df[feature].value_counts(normalize=True) * 100
    original_counts = original_counts.sort_index()
    
    # Calculate percentages for synthetic data
    synthetic_counts = synthetic_df[feature].value_counts(normalize=True) * 100
    synthetic_counts = synthetic_counts.sort_index()
    
    # Plot original data percentages
    sns.barplot(x=original_counts.index, y=original_counts.values, ax=axes[0])
    axes[0].set_title(f'Original Data: Distribution of {feature} (Percentage)')
    axes[0].set_xlabel(feature)
    axes[0].set_ylabel('Percentage')
    
    # Plot synthetic data percentages
    sns.barplot(x=synthetic_counts.index, y=synthetic_counts.values, ax=axes[1])
    axes[1].set_title(f'Synthetic Data: Distribution of {feature} (Percentage)')
    axes[1].set_xlabel(feature)
    axes[1].set_ylabel('Percentage')
    
    plt.show()
```

```python
# Plot distributions for each feature side-by-side
for feature in df_og.columns:
    if feature not in ['id']:
        plot_side_by_side(df_og, df_synth, feature)
```
